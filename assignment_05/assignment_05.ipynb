## Selecting only custom columns

```python
df1 = df[["UNITID", "INSTNM", "CITY", "STABBR", "ZIP", "ADM_RATE", "UGDS", "TUITIONFEE_IN", "MN_EARN_WNE_P10"]]
df1
```

# **Description of each column**



### **UNITID** is Unit ID for institution. These ids are given in academic year 2017-2018 which was reported in IPEDS DCY2017-2018



### **INSTNM** is the name of the institition





### **CITY** column is the city of the institution

### **STABBR** column is the State Abbrevation of the institution

### **ZIP** column is the zip code of the institution

### **ADM_RATE** column is the admisssion rate of the institution. Colleges report admission rate to IPEDS which is defined as number of admitted students to number of students applied to institution.



### **UGDS** column is the number of undergraduate students. This column reports number of students enrolled and reported in the IPEDS

### **TUITIONFEE_IN** column is the In-state tution and fee of the institution 

### **MN_EARN_WNE_P10** column is the mean of the earning of students who are employed
